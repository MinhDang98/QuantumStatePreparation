```python
from qiskit import *
import matplotlib.pyplot as plt
%matplotlib inline
from qiskit.visualization import plot_histogram

qc = QuantumCircuit(2)
# The CNOT looks at two input bits to see whether they are the same or different. 
# Next, it overwrites the target qubit with the answer. 
# The target becomes 0 if they are the same, and 1 if they are different.
qc.cx(0,1)
qc.draw(output='mpl')
```

```python
qc = QuantumCircuit(4,2)
# encode inputs in qubits 0 and 1
qc.x(0) #flips 1st qubit to 1
qc.x(1) #flips 2nd qubit to 1
qc.draw(output='mpl')
```

```python
# use cnots to write the XOR of the inputs on qubit 2
qc.cx(0,2)
qc.cx(1,2)
qc.draw(output='mpl')
```

```python
# use ccx to write the AND of the inputs on qubit 3
qc.ccx(0,1,3)
qc.draw(output='mpl')
```

```python
# extract outputs
qc.measure(2,0) # extract XOR value
qc.measure(3,1) # extract AND value

qc.draw(output='mpl')
```

```python
# Get the last instruction from the circuit data
last_instruction = qc.data[-1]
print(last_instruction)

last_gate_operation = last_instruction.operation
print(last_instruction)

# Correct way to get the qubit indices
last_gate_qubits = [qc.find_bit(q).index for q in last_instruction.qubits]
print(last_gate_qubits)
```
